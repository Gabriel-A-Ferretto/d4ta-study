### Listas

```python
usersLista = ["Gabriel", "Rodrigo", "Julio", 23, 21, 30] # lsita criada [ ]
print(usersLista)
```

```python
listaConvidados = [
    "Paulo",
    "Jose",
    "João",
    "Rogerio"
]

print(listaConvidados)
```

```python
# ADICIONANDO VALOR A UMA LISTA 

vendedores = []

novoVendedor = "Wagner"

# adicionadno valor a uma lista atraves de uma variavel
vendedores.append(novoVendedor) # O metodo -> append() adiciona um valor a lsita
print(f'Estamos com os seguintes vendedores {vendedores} \n')

lista_de_novos_vendedores = [
    "Bilbo Bolseiro",
    "Batman",
    "Demolidor",
    "Lanterna Verde",
    "Mulher Maravilha"
]

vendedores.append('Thor') # adicionando o valor direto na lista

vendedores.append(lista_de_novos_vendedores)
print(f'A nossa nova lista de vendedores foi atualizada. \n CONFIRA A LISTA DOS NOVOS VENDEDORES: \n {vendedores}')



```

```python
# LISTAS  COM INPUTS

addVendedor = str(input('Digite o nome do vendedor que você deseja adicionar a lista: '))
lista_de_novos_vendedores.append(addVendedor)
print(f'O nome do vendedor adicionado foi: {addVendedor} \n')
print(f'A nossa nova lista de vendedores foi atualizada. \n CONFIRA A LISTA DOS NOVOS VENDEDORES: \n {vendedores}')
```

**APAGANDO DADOS DE UMA LISTA**

```python
vendedores.clear() # limpa a lista toda
print(vendedores)
```

```python
listaSuper =  ['Wagner', 'Thor', 'Bilbo Bolseiro', 'Batman', 'Demolidor', 'Lanterna Verde', 'Mulher Maravilha', 'Gabriel']
print(listaSuper)

# remove() busca o nome do indice e não a poisição
listaSuper.remove('Wagner') # removendo um valor especifico 
print(listaSuper)
```

**POSIÇÃO DE UM ITEM DA LISTA**


```python
clientesLoja = [
    "McDonald's",
    "Burger King",
    "Subway",
    "Starbucks",
    "KFC",
    "Domino's Pizza",
    "Pizza Hut",
    "Havan",
    "Magazine Luiza",
    "Casas Bahia",
    "Carrefour",
    "Extra"
]

print(clientesLoja)

primeiroClinte = clientesLoja[0] # buscando o indice 0 que no caso é o primeiro a contagem de uma lista começa de 0,1,2... 
print(primeiroClinte)

clienteNcinco = clientesLoja[5]
print(clienteNcinco)

print(f'\n Os clientes que foram mias requisitados foram {primeiroClinte} e {clienteNcinco}')

procuraCliente = 'Extra'
poiscao = clientesLoja.index(procuraCliente) # index() esta buscando o nome e não o indice dentro da lista 
print(f'\n A posição do cliente {procuraCliente.upper()} dentro da nossa lista é {poiscao}')
```
