### 3.7 match / case
> `match_e_case.py` — Estrutura match/case (Python 3.10+) para comparar um valor único.

```python
protocolo = "TCP"

match protocolo:
    case "TCP":
        print("Conexão orientada")
    case "UDP":
        print("Sem conexão")
    case _:
        print("Desconhecido")
```

### 3.8 Analisador de Eventos (função + condicionais)
> `analisaEvento.py` — Função que analisa dicionários de eventos de segurança.

```python
def analisar_evento(evento):
    nivel      = evento.get("nivel")
    origem     = evento.get("origem")
    tentativas = evento.get("tentativas", 0)

    if nivel == "CRITICAL":
        print(f"[ALERTA CRÍTICO] Origem: {origem} — escalar imediatamente")
    elif nivel == "HIGH" and tentativas > 10:
        print(f"[POSSÍVEL BRUTE FORCE] {tentativas} tentativas de {origem}")
    elif nivel == "HIGH":
        print(f"[ALERTA ALTO] Investigar: {origem}")
    elif nivel in ("MEDIUM", "LOW"):
        print(f"[INFO] Registrado — nível {nivel} de {origem}")
    else:
        print("[DESCONHECIDO] Evento sem nível definido")

# Testes
analisar_evento({"nivel": "HIGH", "origem": "10.0.0.5", "tentativas": 15})
analisar_evento({"nivel": "CRITICAL", "origem": "192.168.1.1"})
analisar_evento({"nivel": "LOW", "origem": "172.16.0.2", "tentativas": 2})
```

### 4.9 `while` com `threading` — Monitoramento de serviço
> `monitoramento_While.py` — Simulação de monitoramento com menu interativo e thread em paralelo.

> ⚠️ Este exemplo usa `input()` e `time.sleep()`. No Jupyter, execute célula por célula e use o botão ⏹ para interromper.

```python
import time
import threading
import random

# Simulação de serviço
def checar_servico():
    return random.choice(["online", "online", "online", "offline"])

def escanear():
    print("🔎 Escaneando sistema...")
    time.sleep(2)
    print("✔ Escaneamento concluído")

def gerar_relatorio():
    print("📄 Gerando relatório...")
    time.sleep(2)
    print("✔ Relatório pronto")

def monitorar():
    while True:
        status = checar_servico()
        if status == "offline":
            print("\n🚨 Serviço caiu! ALERTA!")
            break
        print("✅ OK — aguardando...")
        time.sleep(5)

# Thread de monitoramento em segundo plano
monitor_thread = threading.Thread(target=monitorar, daemon=True)
monitor_thread.start()

# Menu interativo
while True:
    try:
        opcao = input("\n[1] Escanear  [2] Relatório  [0] Sair\n> ")
        if opcao == "0":
            print("Saindo...")
            break
        elif opcao == "1":
            escanear()
        elif opcao == "2":
            gerar_relatorio()
        else:
            print("❌ Opção inválida")
    except KeyboardInterrupt:
        print("\nEncerrando...")
        break
```

### 5.3 Métodos de Lista
> `metodosLista.py` — Arquivo originalmente vazio; exemplos dos principais métodos de lista em Python.

```python
frutas = ['maçã', 'banana', 'laranja']

frutas.append('uva')          # adiciona ao final
print('append:', frutas)

frutas.insert(1, 'kiwi')      # insere na posição 1
print('insert:', frutas)

frutas.remove('banana')       # remove pelo valor
print('remove:', frutas)

ultimo = frutas.pop()         # remove e retorna o último
print('pop:', ultimo, '→ lista:', frutas)

print('index:', frutas.index('maçã'))  # índice do elemento
print('count:', frutas.count('kiwi'))  # quantas vezes aparece

frutas.sort()                 # ordena in-place
print('sort:', frutas)

frutas.reverse()              # inverte in-place
print('reverse:', frutas)

print('len:', len(frutas))    # comprimento
```
